```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

# Visualize and analyze the minimum, maximum and median income

```python
y=list(df.income)
plt.boxplot(y)
plt.show()
```

# Find out the no. of days of reduced activity of male and female separately due to illness

```python
df.groupby(['sex','actdays']).sum()
```

# Visualize if there is any missing values in the dataset

```python
sns.heatmap(df.isnull(),cbar=False,cmap='viridis')
plt.show()
```

# Find the correlation between different variables

```python
num_df = df.select_dtypes(include=[np.number])
plt.figure(figsize=(10,10))
sns.heatmap(num_df.corr(),cbar=True,annot=True,cmap="Blues")
plt.show()
```

# How income affects the no. of visit to the hospital

```python
plt.figure(figsize=(10,10))
plt.scatter(x='income',y='hospadmi',data=num_df)
plt.ylabel('visits')
plt.xlabel('income')
```

# to visualize the no. of males and females using illness

```python
sns.histplot(df.sex,bins=2)
```

# percentage of people getting govt. insurance

```python
print(df['insurance'].unique())
```

```python
low_income = df[df['insurance'] == 'freepor'].shape[0]
old_age    = df[df['insurance'] == 'freerepa'].shape[0]
private    = df[df['insurance'] == 'levyplus'].shape[0]
values = [low_income, old_age, private]
labels = ['Govt (Low Income)', 'Govt (Old Age/Disability)', 'Private Insurance']

print("Values:", values)

if sum(values) > 0:
    plt.figure(figsize=(6, 6))
    plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)
    plt.title("Percentage of People by Type of Health Insurance")
    plt.show()
else:
    print("No valid data to plot.")
```

# Plotting bar chart for the reduced number of days due to illness based on gender

```python
db = df.groupby('sex')['actdays'].sum().to_frame().reset_index()

# Creating the bar chart
plt.barh(db['sex'], db['actdays'], color=['cornflowerblue', 'lightseagreen'])

# Adding the aesthetics
plt.title('Bar Chart')
plt.xlabel('reduced activity')
plt.ylabel('gender')

# Show the plot
plt.show()
```
